```python
horror_df = final_merge_df[final_merge_df['genres'].str.contains('Horror')]
horror_df = horror_df.drop(columns = 'Is_Horror')
horror_df
```

```python
list_to_drop_from_reg = ['Worldwide Gross:Production Ratio','Spring', 'Summer', 'Fall', 'Winter','30 min','1 hour','1.5 hours','2 hours' ,'2.5 hours', '3 hours', 'doy', 'season','Profit:Production Ratio', 'time_grouping', 'movie_id', 'start_year', 'primary_title', 'original_title', 'genres', 'runtime_minutes', 'director_name', 'writer_name', 'actors', 'id', 'release_date', 'movie', 'year', 'Profit', 'production_budget', 'domestic_gross', 'worldwide_gross']
X2 = horror_df.drop(columns = list_to_drop_from_reg + genre_list)
y2 = horror_df['Profit:Production Ratio']
X2
X_augmented2 = sm.add_constant(X2)
results2 = sm.OLS(y2, X_augmented2).fit()
results2.summary()
```

```python
weights = results2.params
weights
```

```python
#plotting the effect of Month on ppr from regression
plt.figure(figsize=(10,5))
sns.barplot(y = weights[1::].index, x = weights[1::].values)
plt.title("Expected PPR by Month")
plt.ylabel('Month')
plt.xlabel('Expected Change in PPR')
plt.show()
```

```python
# standardizing results
X_standard = X2.apply(lambda x: (x-x.mean())/ x.std())

#adding constant to standardized results so regression works and creating the results_standardized object
X_standard_aug = sm.add_constant(X_standard)
results_standardized = sm.OLS(y2, X_standard_aug).fit()
```

```python
#getting the weights
weights_standardized = results_standardized.params
weights_standardized
```

```python
#plotting the effect of genre on ppr from regression w/ standardization
plt.figure(figsize=(10,5))
sns.barplot(y = weights_standardized[1::].index, x = weights_standardized[1::].values)
plt.title("Expected PPR by Month")
plt.ylabel('Month')
plt.xlabel('Expected Change in PPR')
plt.savefig('../images/month_plot.png')
plt.show()
```
