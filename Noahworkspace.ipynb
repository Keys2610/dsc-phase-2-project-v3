```python
merged_df.info()
```

```python
list_to_drop_from_reg = ['Profit:Production Ratio', 'movie_id', 'start_year', 'primary_title', 'original_title', 'genres', 'runtime_minutes', 'director_name', 'writer_name', 'actors', 'id', 'release_date', 'movie', 'year', 'Profit', 'production_budget', 'domestic_gross', 'worldwide_gross']
X = merged_df.drop(columns = list_to_drop_from_reg)
y = merged_df['Profit:Production Ratio']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 8))
sns.heatmap(X.corr(), annot=True)
plt.title('genre regression heatmap')
plt.show()
```

```python
import numpy as np
mask = np.triu(np.ones_like(X.corr(), dtype=bool))
heatmap = sns.heatmap(X.corr(), mask=mask, vmin=-1, vmax=1, annot=True)
heatmap.set_title('lower triangular genre heatmap', fontdict={'fontsize':18}, pad=14);
```

```python
X_augmented = sm.add_constant(X)
X_augmented
results = sm.OLS(y, X_augmented).fit()
print(results)
```

```python
X_standard = X.apply(lambda x: (x-x.mean())/ x.std())
```

```python
X_standard_aug = sm.add_constant(X_standard)
results_standardized = sm.OLS(y, X_standard_aug).fit()
```

```python
weights = results_standardized.params
weights
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y = weights[1::].index, x = weights[1::].values)
plt.title("genre regression")
plt.ylabel('regression')
plt.show()
```

```python
results_standardized.summary()
```
