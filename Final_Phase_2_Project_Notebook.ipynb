## Cleaning

Code and code comments in data_cleaning.ipynb

Once we had our joined dataframe we went into our data cleaning. For our cleaning there were only 3 cleaning steps that we took and I will explain and justify them here. Our first step was to remove all movies that grossed 0 dollars at the box office. The reason for this is because we thought that many of these movies were the victim of tax loss harvesting where a studio makes a movie, does test screenings, and if it isn't recieved well then the studio will decide not to release it and just write it off on their taxes for the year as a failed investment for example: the new [catwoman movie](https://finance.yahoo.com/news/why-did-warner-bros-kill-202941101.html?guccounter=1) that Warner Bros. decided to not release and instead write it off as a loss. The second group of movies that we dropped from our dataset is movies that were made before the year 2000. The reasoning behind this is we wanted to only capture movies that are relevant to the current day movie market as we are recommending movies for a studio to produce for current day. We had 2 ideas for what our cutoff should be for the date either 2008 or 2000. At first we thought that 2008 would be the best cutoff as thats when the first [Iron Man](https://www.imdb.com/title/tt0371746/) movie came out, which many argue was the start of the "cinematic universe" trend in hollywood which is what currently dominates the box office. However, upon further inspection, this trend actually dates further back to the year 2000 as there were many movies that came out at this time that set up super franchises in the later 21st century such as [The Fast and the Furious](https://www.imdb.com/title/tt0232500/), the first [Lord of the Rings](https://www.imdb.com/title/tt0120737/) film, and [Shrek](https://www.imdb.com/title/tt0126029/). The final cleaning step we took was to remove movies that had a budget under 1,000,000 dollars as we would consider these more personal film projects or small indie films than movies that are relevant to recommending to a full scale movie studio.

We did a little bit of minor cleaning that is worth mentioning as well. We converted all of our budget and revenue data from strings to intergers so that we could use them in analysis.
